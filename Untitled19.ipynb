```python
import gym, math
import numpy as np
import matplotlib.pyplot as plt

class mountaincar_driver_qlearning:
    def __init__(self, alpha, gamma, epsilon, min_epsilon):
        print("------------------------------")
        print("Initializing Training...")
        print("Learning Rate: {}".format(alpha))
        print("Discount Rate: {}".format(gamma))
        print("Initial Exploration Rate: {}".format(epsilon))
        print("------------------------------")
        self.qtable = np.random.uniform(low=-1, high=1, size=(19, 15, 3))
        self.alpha = alpha
        self.gamma = gamma
        self.epsilon = epsilon
        self.min_epsilon = min_epsilon
        self.env = gym.make('MountainCar-v0')
        self.avg_reward_list = np.array([])
        self.avg_reward_header_list = np.array([])

    # multiply to make values whole numbers then round up
    def descretize_state(self, state):
        position = math.ceil(state[0] * 10)
        velocity = math.ceil(state[1] * 100)
        return [position, velocity]

    def train(self, episodes):
        ave_reward = 0
        epsilon_reduction = (self.epsilon - self.min_epsilon) / episodes
        convergence_time = 0
        flag = False
        success = 0
        for i in range(episodes):
            # Initialize params
            done = False
            reward = 0
            total_reward = 0
            state = self.descretize_state(self.env.reset())
            # Run episode
            while not done:

                # Determine action
                if np.random.random() < self.epsilon:  # random action
                    action = self.env.action_space.sample()
                else:
                    action = np.argmax(self.qtable[state[0], state[1]])

                # Get reward and Next state
                next_state, reward, done, _ = self.env.step(action)
                next_state = self.descretize_state(next_state)

                # update qtable
                old_value = self.qtable[state[0], state[1], action]
                nextmax = np.max(self.qtable[next_state[0], next_state[1]])
                self.qtable[state[0], state[1], action] = (1 - self.alpha) * old_value + self.alpha * (
                            reward + self.gamma * nextmax)

                # update variables
                total_reward += reward
                state = next_state

            # Decay Epsilon
            if self.epsilon > self.min_epsilon:
                self.epsilon -= epsilon_reduction

            # calculate accuracy
            if total_reward > -200.0 and flag == True:
                success += 1

            if total_reward > -200.0 and flag == False:
                convergence_time = i
                flag = True
            # Keep track of average reward over 100 episodes
            ave_reward += total_reward
            if (i + 1) % 100 == 0:
                average = ave_reward / 100
                ave_reward = 0
                print("Episode {}.  Average Reward: {}".format(i + 1, average))
                self.avg_reward_list = np.append(self.avg_reward_list, average)
                self.avg_reward_header_list = np.append(self.avg_reward_header_list, i+1)



        accuracy = success / episodes * 100
        print("Accuracy: {}".format(accuracy))
        print("Convergence Time: {}".format(convergence_time))

        self.plot_average(self.avg_reward_header_list, self.avg_reward_list)


    def plot_average(self,x,y):
        plt.title("Average Reward")
        plt.plot(x, y, color="red")

        plt.show()

    def display(self, episodes):
        for _ in range(episodes):
            state = self.descretize_state(self.env.reset())
            done = False
            while not done:
                self.env.render()
                action = np.argmax(self.qtable[state[0], state[1]])
                next_state, _, done, _ = self.env.step(action)
                next_state = self.descretize_state(next_state)
                state = next_state
        self.env.close()

class mountaincar_driver_montecarlo:
    def __init__(self):
        print("------------------------------")
        print("Initializing Training...")
        print("------------------------------")
        self.policy = [[None for _ in range(0, 15)] for _ in range(0, 19)]  # Policy for agent defined by state space
        self.q = [[[0 for _ in range(0, 3)] for _ in range(0, 15)] for _ in range(0, 19)]  # State-action-value matrix
        self.ep = 0  # Episode
        self.max = 2000  # Max episode
        self.latest_states = [[0 for _ in range(0, 3)] for _ in range(0, 1000)]  # States accessed in the latest episode
        self.epsilon = 0  # Mutation rate
        self.reward = []  # Episode reward matrix
        self.mutated = []
        self.env = gym.make('MountainCar-v0')
        self.env._max_episode_steps = 200
        self.avg_reward_list = np.array([])
        self.avg_reward_header_list = np.array([])

    # multiply to make values whole numbers then round up
    def state_formatter(self, state):
        """
        formats state space to positive wole numbers
        """

        # Round state space to tenths place to form finite values

        state[0] = state[0].round(1)
        state[1] = state[1].round(2)

        # Make numbers whole

        state[0] *= 10
        state[1] *= 100

        # Make numbers whole

        state[0] += 12
        state[1] += 7

        # Change data type

        integer = [0, 0]
        integer[0] = int(state[0])
        integer[1] = int(state[1])

        return integer

    def calc_q(self, latest_states):
        """
        updates state values
        """
        for i in range(0, 19):
            for p in range(0, 15):
                for u in range(1000):

                    # If value occurs in last episode, update it according to episode reward

                    if [i, p, self.policy[i][p]] == [latest_states[u][0], latest_states[u][1], latest_states[u][2]]:
                        self.q[i][p][latest_states[u][2]] = (self.q[i][p][latest_states[u][2]] / 2) + self.reward[self.ep]

    def greedy(self):
        """
        calculates optimal policy
        """
        for i in range(0, 19):
            for p in range(0, 15):

                # If state is unexplored, do not calculate optimal action
                # Else, update the policy accoring to epsilon

                if self.q[i][p][0] == 0 and self.q[i][p][1] == 0 and self.q[i][p][2] == 0:
                    self.policy[i][p] = None
                else:
                    if self.epsilon > np.random.uniform(0, 1):
                        self.policy[i][p] = self.q[i][p].index(np.max(self.q[i][p]))
                    else:
                        self.policy[i][p] = np.random.randint(0, 2)

    def calc_reward(self):
        """
        calculates episode reward by finding the maximum X value
        """
        arr = []
        for i in range(1000):
            arr.append(self.latest_states[i][0])
        return np.max(arr)

    def train(self):
        convergence_time = 0
        ave_reward = 0
        success = 0
        while self.ep < self.max:
            self.epsilon += 0.001  # Update epsilon for convergence
            count = 0  # Reset frame counter
            formatted_state = self.state_formatter(self.env.reset())  # Reset environment
            total_reward = 0
            flag = False
            end = False
            while not end:

                # If state is unexplored, use uniform random action

                if self.policy[formatted_state[0]][formatted_state[1]] == None:
                    self.policy[formatted_state[0]][formatted_state[1]] = np.random.randint(0, 2)

                # Step forward in environment with action and receive new state

                state, reward, end, info = self.env.step(self.policy[formatted_state[0]][formatted_state[1]])

                # Set local state space

                self.latest_states[count][0] = formatted_state[0]
                self.latest_states[count][1] = formatted_state[1]
                self.latest_states[count][2] = self.policy[formatted_state[0]][formatted_state[1]]

                # Format new state space

                formatted_state = self.state_formatter(state)
                total_reward += reward
                count += 1

            self.reward.append(self.calc_reward())  # Append reward matrix for graph
            self.calc_q(self.latest_states)  # Calculate Q
            self.greedy()  # e-greedy function

            # calculate accuracy
            if total_reward > -200.0 and flag == True:
                success += 1

            if total_reward > -200.0 and flag == False:
                convergence_time = self.ep
                flag = True

            # Keep track of average reward over 100 episodes
            ave_reward += total_reward
            if (self.ep + 1) % 100 == 0:
                average = ave_reward / 100
                ave_reward = 0
                print("Episode {}.  Average Reward: {}".format(self.ep + 1, average))
                self.avg_reward_list = np.append(self.avg_reward_list, average)
                self.avg_reward_header_list = np.append(self.avg_reward_header_list, self.ep + 1)

            self.ep += 1

        accuracy = success / self.ep * 100
        print("Accuracy: {}".format(accuracy))
        print("Convergence Time: {}".format(convergence_time))

        self.plot_average(self.avg_reward_header_list, self.avg_reward_list)

    def plot_average(self, x, y):
        plt.title("Average Reward")
        plt.plot(x, y, color="red")

        plt.show()

    def display(self, episodes):
        for _ in range(episodes):
            state = self.state_formatter(self.env.reset())
            done = False
            while not done:
                self.env.render()
                action = self.policy[state[0]][state[1]]
                next_state, _, done, _ = self.env.step(action)
                next_state = self.state_formatter(next_state)
                state = next_state
        self.env.close()

class mountaincar_driver_sarsa:
    def __init__(self):
        print("------------------------------")
        print("Initializing Training...")
        print("------------------------------")
        self.env = gym.make('MountainCar-v0')
        self.avg_reward_list = np.array([])
        self.avg_reward_header_list = np.array([])

        # Discritize observation and action space in bins.
        self.pos_space = np.linspace(-1.2, 0.6, 18)
        self.vel_space = np.linspace(-0.07, 0.07, 28)
        self.env._max_episode_steps = 200

        # Create an empty Q-table
        self.Q = self.createEmptyQTable()

        # Hyperparameters
        self.alpha = 0.1  # Learning Rate
        self.gamma = 0.9  # Discount Factor
        self.epsilon = 1  # e-Greedy
        self.episodes = 2000  # number of episodes

        self.score = 0
        # Variable to keep track of the total score obtained at each episode
        self.total_score = np.zeros(self.episodes)

    # given observation, returns what bin
    def getState(self, observation):
        pos, vel = observation
        pos_bin = np.digitize(pos, self.pos_space)
        vel_bin = np.digitize(vel, self.vel_space)

        return (pos_bin, vel_bin)

    # Creates a new empty Q-table for this environment
    def createEmptyQTable(self):
        states = []
        for pos in range(len(self.pos_space) + 1):
            for vel in range(len(self.vel_space) + 1):
                states.append((pos, vel))

        Q = {}
        for state in states:
            for action in range(self.env.action_space.n):
                Q[state, action] = 0
        return Q

    # Given a state and a set of actions
    # returns action that has the highest Q-value
    def maxAction(self, Q, state, actions=[0, 1, 2]):
        values = np.array([Q[state, a] for a in actions])
        action = np.argmax(values)
        return action

    def train(self):
        convergence_time = 0
        success = 0
        flag = False
        ave_reward = 0
        for i in range(self.episodes):
            observation = self.env.reset()
            state = self.getState(observation)

            # e-Greedy strategy
            # Explore random action with probability epsilon
            if np.random.uniform(0, 1) < self.epsilon:
                action = self.env.action_space.sample()
            # Take best action with probability 1-epsilon
            else:
                action = self.maxAction(self.Q, state)

            total_reward = 0
            done = False

            while not done:
                # Take action and observe next state
                next_observation, reward, done, info = self.env.step(action)
                next_state = self.getState(next_observation)

                # Get next action following e-Greedy policy
                if np.random.uniform(0, 1) < self.epsilon:
                    next_action = self.env.action_space.sample()
                else:
                    next_action = self.maxAction(self.Q, next_state)

                # Add reward to the score of the episode
                total_reward += reward

                # Update Q value for state and action given the bellman equation
                self.Q[state, action] = self.Q[state, action] + self.alpha * (
                            reward + self.gamma * self.Q[next_state, next_action] - self.Q[state, action])

                # Move to next state, and next action
                state, action = next_state, next_action

                # calculate accuracy
            if total_reward > -200.0 and flag == True:
                success += 1

            if total_reward > -200.0 and flag == False:
                convergence_time = i
                flag = True

            # Keep track of average reward over 100 episodes
            ave_reward += total_reward
            if (i + 1) % 100 == 0:
                average = ave_reward / 100
                ave_reward = 0
                print("Episode {}.  Average Reward: {}".format(i + 1, average))
                self.avg_reward_list = np.append(self.avg_reward_list, average)
                self.avg_reward_header_list = np.append(self.avg_reward_header_list, i + 1)

            self.total_score[i] = total_reward
            self.epsilon = self.epsilon - 2 / self.episodes if self.epsilon > 0.01 else 0.01

        accuracy = success/self.episodes * 100
        print("Accuracy: {}".format(accuracy))
        print("Convergence Time: {}".format(convergence_time))

        self.plot_average(self.avg_reward_header_list, self.avg_reward_list)

    def plot_average(self, x, y):
        plt.title("Average Reward")
        plt.plot(x, y, color="red")

        plt.show()

    def display(self, episodes):
        for episode in range(10):
            done = False
            observation = self.env.reset()
            state = self.getState(observation)
            # While the car don't reach the goal or number of steps < 200
            while not done:
                self.env.render()
                # Take the best action for that state given trained values
                action = self.maxAction(self.Q, state)
                observation, reward, done, info = self.env.step(action)
                # Go to next state
                state = self.getState(observation)

        self.env.close()

# montecarlo
dEarnhardt = mountaincar_driver_montecarlo()
dEarnhardt.train()
#dEarnhardt.display(3)

# qlearning
alpha = .2  # learning rate
gamma = .9  # future reward discount
epsilon = .8  # random action rate
min_epsilon = 0
episodes = 2000

dEarnhardt = mountaincar_driver_qlearning(alpha, gamma, epsilon, min_epsilon)
dEarnhardt.train(episodes)
#dEarnhardt.display(10)

# SARSA
dEarnhardt = mountaincar_driver_sarsa()
dEarnhardt.train()
#dEarnhardt.display(10)
```
